# D1 perfomance (20 random_states)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = '/Users/kamme/Documents/study/pythonProject/RandomForest/RF_D1_results.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataframe to understand its structure
print(df.head())

# Extract metrics
metrics = ['Accuracy', 'Recall', 'Precision', 'F1-Score']

# Calculate the average for each metric
average_metrics = df[metrics].mean()
max_metrics = df[metrics].max()
min_metrics = df[metrics].min()
print(f"\nAverage\n{average_metrics}\n")
print(f"Max\n{max_metrics}\n")
print(f"Min\n{min_metrics}\n")

# Plot the average line (dashed line)
plt.plot(metrics, average_metrics, label='Average', color='black', linestyle='dashed')

# Plot for each seed
for index, row in df.iterrows():
    plt.plot(metrics, row[metrics], label=f'Seed {row["Seed"]}')

plt.title('Performance of RF on D1 (different seeds)')
plt.legend(loc='upper left')
plt.xlabel('Metrics')
plt.ylabel('Scores')
plt.grid(True)

# Place the legend outside the plot
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# Adjust the plot size to make room for the legend
plt.gcf().set_size_inches(10, 6)
#plt.savefig("RF_D1_20.png", bbox_inches='tight')
plt.show()
```

Confidence Interval

The graph suggests that the Random Forest model might have achieved higher average Recall on Dataset D2 compared to others. However, to confirm this, we would need to perform a statistical test (e.g., ANOVA) to assess if the observed difference is statistically significant. Additionally, Dataset D4 seems to have the widest confidence intervals for most metrics, indicating higher uncertainty in its performance estimates.

```python
import numpy as np
from scipy.stats import sem
import matplotlib.pyplot as plt

# Sample data (replace with your actual data)
metrics = ['Recall', 'Accuracy', 'Precision', 'F1']

# Data for Model RF on 4 datasets (replace with actual values)
model_RF_D1 = [0.568518, 0.758117, 0.675497, 0.614935]
model_RF_D2 = [0.588876, 0.764286, 0.681032, 0.629929]
model_RF_D3 = [0.576068, 0.768354, 0.681692, 0.621430]
model_RF_D4 = [0.515542, 0.765068, 0.699275, 0.586703]

# Convert to numpy arrays for easier calculation
data = np.array([model_RF_D1, model_RF_D2, model_RF_D3, model_RF_D4])

# Calculate mean and standard error for each metric
means = np.mean(data, axis=0)
errors = sem(data, axis=0)  # Standard error of the mean

# Setting the positions and width for the bars
x = np.arange(len(metrics))
width = 0.35

# Creating the bar plot with error bars (95% confidence intervals)
fig, ax = plt.subplots()
bars = ax.bar(x, means, width, yerr=errors * 1.96, capsize=5,ecolor='red', label='Model RF')

# Adding labels, title, and legend
ax.set_xlabel('Metrics', fontsize=14)
ax.set_ylabel('Scores', fontsize=14)
ax.set_title('Average Performance with 95% Confidence Intervals', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(metrics, fontsize=12)
ax.legend(fontsize=12)

# Display the bar plot
plt.savefig("RF_confidence.png")
plt.show()
```
