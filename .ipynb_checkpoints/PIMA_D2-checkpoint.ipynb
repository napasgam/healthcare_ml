Impute 'zero' value by means

```python
# Copy the first 8 columns
rows_to_impute = df.iloc[:, :8].copy()

# Replace zero values with NaN
rows_to_impute.replace(0, np.nan, inplace=True)

# Display the dataset after replacing zeros with NaN
print("\nDataset after replacing zeros with NaN:")
print(rows_to_impute)
```

```python
# Calculate means of the columns before imputation
column_means_before_imputation = rows_to_impute.mean()
print("\nColumn means before imputation:")
print(column_means_before_imputation)
```

Split Data

```python
x = df.iloc[:,0:8] #select all rows, select columns 0 to 8
y = df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
X_train, x_test, y_train, y_test = train_test_split(x ,y , test_size=0.3, random_state=0)
# Split train and test into 70:30
```

##Random Forest

Baseline model with default parameters

```python
%%time
from sklearn.ensemble import RandomForestClassifier

# Create a RandomForestClassifier object
rf_classifier = RandomForestClassifier(n_estimators=50, random_state=1)

# Fit the model on training data
rf_classifier.fit(X_train, y_train)

# Make predictions on new data
y_pred = rf_classifier.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
# Accuracy = TP + TN / (TP + TN + FP + FN)
# Precision = TP / (TP + FP)
# Recall = TP / (TP +  FN)  Also known as sensitivity, or True Positive Rate
# F1 = 2 * Precision * Recall / (Precision + Recall)

print('Accuracy : {:.2f}' .format(accuracy_score(y_pred,y_test)))
print('Precision: {:.2f}' .format(precision_score(y_test,y_pred)))
print('Recall   : {:.2f}' .format(recall_score(y_test,y_pred)))
print('F1       : {:.2f}' .format(f1_score(y_pred,y_test)))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

Grid Search

```python
parameters ={
      'criterion' : ['gini', 'entropy'],
      'max_features': ['sqrt'],
      'random_state': range(1, 51, 5),
      'min_samples_leaf': range(1, 51, 2),
      'n_estimators': range(1, 51, 5)}
```

```python
%%time
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
# Create a RandomForestClassifier object
clf = RandomForestClassifier()
rf_classifier = GridSearchCV(clf, parameters, cv=4, scoring='accuracy',n_jobs=-1 ,error_score='raise')

# Fit the model on training data
rf_classifier.fit(X_train, y_train)

# Make predictions on new data
y_pred = rf_classifier.best_estimator_.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.metrics import accuracy_score

print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
acc2 = accuracy_score(y_test,y_pred)
```
